### NBA if Las Vegas and Seattle expansion occur

```python
from sportwarz import Team, Coordinates

team_LV = Team(
    name = "Las Vegas Scorpions" ,
    L = 2.0,
    S = 3.0,
    venue = "T-Mobile Arena",
    state = "Nevada",
    color = "red",
    coordinates = Coordinates(
        lat = 36.102504,
        lon = -115.178729
))

team_SEA = Team(
    name = "Seattle Supersonics" ,
    L = 2.0,
    S = 3.0,
    venue = "Climate Pledge Arena",
    state = "Washington",
    color = "darkgreen",
    coordinates = Coordinates(
        lat = 47.622548,
        lon = -122.354813
))

leagues_modified = leagues_model.copy_with_just_league(league_name)

leagues_modified.add_teams(league_name, [team_LV, team_SEA])

leagues_modified.heatmap_counties(league_name) 
leagues_modified.render_map(league_name)
```

```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.0f' % x)
pd.set_option('display.max_rows', 256)
pd.set_option('display.width', 256)

leagues_model.show_pre_post_merged_results(
    league_name,
    leagues_modified
)
```
