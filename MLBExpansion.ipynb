```python
#Complete move of A's to Vegas
from sportwarz import replace_team, add_team, Team, Coordinates

league_name = "MLB"    

team_LVAs = Team(
    name = "Las Vegas A's" ,
    L = 2.0,
    S = 3.0,
    venue = "Fisher's Armadillo",
    state = "Nevada",
    color = "darkgreen",
    coordinates = Coordinates(
        lat = 36.099444,
        lon = -115.170000
))

leagues = replace_team(co_data_frame, league_name, "Athletics", team_LVAs)

reset_county_styles(counties_geojson)
heatmap_counties(leagues, counties_geojson, league_name) 

render_map(leagues, league_name, counties_geojson, co_data_frame)
```

```python
from sportwarz import Team, Coordinates, league_teams_sums, add_team, reset_county_styles, heatmap_counties

league_name = "MLB"

team_SLC = Team(
    name = "Salt Lake Bees" ,
    L = 2.0,
    S = 3.0,
    venue = "Bees Knees Field",
    state = "Utah",
    color = "yellow",
    coordinates = Coordinates(
        lat =  40.5497,
        lon = -112.0225
))

leagues_MLB_wSLC = add_team(co_data_frame, league_name, [team_SLC])

import pandas as pd

pre_sums = league_teams_sums(leagues[league_name])

post_SLC_sums = league_teams_sums(leagues_MLB_wSLC[league_name])

merged = pd.merge(pre_sums, post_SLC_sums, on='team_name', how='right')
merged["share_population_x"].fillna(0, inplace = True) 

pd.set_option('display.float_format', lambda x: '%.0f' % x)

print(merged)

print (merged["share_population_x"].sum())
print (merged["share_population_y"].sum()) 

reset_county_styles(counties_geojson)
heatmap_counties(leagues_MLB_wSLC, counties_geojson, league_name) 

render_map(leagues_MLB_wSLC, league_name, counties_geojson, co_data_frame)
```
