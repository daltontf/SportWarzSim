```python
from sportwarz import league_teams_sums

import pandas as pd

pd.set_option('display.float_format', lambda x: '%.0f' % x)

def show_pre_post_merged_results(league_before, league_after):
    pre_sums = league_teams_sums(league_before)
    post_sums = league_teams_sums(league_after)
    merged = pd.merge(pre_sums, post_sums, on='team_name', how='right')
    merged["share_population_x"].fillna(0, inplace = True) 
    merged["share_population_delta"] = merged["share_population_y"] - merged["share_population_x"]
    print(merged)
    print(f"Sums\t: {merged["share_population_x"].sum()}\t{merged["share_population_y"].sum()}")
    print(f"Sum Delta = {merged["share_population_y"].sum() - merged["share_population_x"].sum()}") 
```

```python
# Complete move of A's to Vegas
from sportwarz import update_team, add_team, Team, Coordinates, reset_county_styles, copy_with_just_league

team_LVAs = Team(
    name = "Las Vegas A's" ,
    L = 5.0,
    S = 1.0,
    venue = "Fisher's Armadillo",
    state = "Nevada",
    coordinates = Coordinates(
        lat = 36.099444,
        lon = -115.170000
))

#don't mutate leagues
leagues_wLAV = copy_with_just_league(leagues, league_name)

update_team(co_data_frame, leagues_wLAV, league_name, "Athletics", team_LVAs)

reset_county_styles(counties_geojson)
heatmap_counties(leagues_wLAV, counties_geojson, league_name) 

render_map(leagues_wLAV, league_name, counties_geojson, co_data_frame)
```

```python
show_pre_post_merged_results(
    leagues[league_name],
    leagues_wLAV[league_name], 
)
```
