### NFL right now

```python
from sportwarz import Simulation

league_name = "NFL"

simulation  = Simulation()
simulation.add_league(league_name)
simulation.render_current_model(league_name)                       
```

### NFL if St. Louis expansion over Jacksonville

```python
from sportwarz import Team, Coordinates

simulation.relocate_team(league_name, "Jacksonville Jaguars", Team(   
    name = "St. Louis Stallions" ,
    venue = "National Car Rental Stadium", 
    state = "Missouri",
    color = "plum",
    coordinates = Coordinates(
        lat =  38.63911966902621,
        lon = -90.182295076939940
)))

simulation.render_current_model(league_name)
```
