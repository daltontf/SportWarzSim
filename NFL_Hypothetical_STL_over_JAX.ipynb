### NFL if 1995 expansion team went to St. Louis intead of Jacksonville

```python
# What if STL got the '95 expansion team instead of JAX
from sportwarz import Team, Coordinates

team_STL = Team(
    name = "St. Louis Stallions" ,
#    L = 3.0,
#    S = 3.0,
    venue = "National Car Rental Stadium", 
    state = "Missouri",
    color = "plum",
    coordinates = Coordinates(
        lat =  38.63911966902621,
        lon = -90.182295076939940
))

leagues_modified = leagues_model.copy_with_just_league(league_name)

leagues_modified.update_team(league_name, "Jacksonville Jaguars", team_STL)

leagues_modified.heatmap_counties(league_name) 
leagues_modified.render_map(league_name)
```

```python
import itables

df = leagues_model.show_pre_post_merged_results(
    league_name,
    leagues_modified
)

itables.show(df, paging=False, pageLength=100 )
```
