### MLB right now

```python
from sportwarz import Simulation

league_name = "MLB"

simulation  = Simulation()
simulation.add_league(league_name)
simulation.render_current_model(league_name)                            
```

### MLB moving Rays to Orlando

```python
from sportwarz import Team, Coordinates

simulation.update_team(league_name, "Tampa Bay Rays", 
  Team(  
    name = "Orlando Dreamers" ,
    color = "black",
    venue = "Dreamers Park",
    state = "Florida",
    coordinates = Coordinates(
        lat = 28.41936717894132, 
        lon = -81.45544944850735
)))

simulation.render_current_model(league_name)
```
