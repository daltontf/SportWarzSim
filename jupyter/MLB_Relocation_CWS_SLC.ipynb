### MLB right now

```python
from sportwarz import Simulation

league_name = "MLB"

simulation  = Simulation()
simulation.add_league(league_name)
simulation.render_current_model(league_name)                            
```

### MLB moving White Sox to Salt Lake City

```python
from sportwarz import Team, Coordinates

simulation.update_team(league_name, "Chicago White Sox", 
  Team(  
    name = "Salt Lake Bees" ,
    venue = "Bees Knees Field",
    state = "Utah",
    color = "yellow",
    coordinates = Coordinates(
        lat =  40.5497,
        lon = -112.0225
)))

simulation.render_current_model(league_name)
```
