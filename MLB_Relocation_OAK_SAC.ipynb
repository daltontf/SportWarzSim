### MLB with Athletics in Oakland

```python
from sportwarz import Simulation, Team, Coordinates

league_name = "MLB"

simulation  = Simulation()
simulation.add_league(league_name)

simulation.update_team(league_name, "Athletics", 
  Team(  
    name = "Oakland A's" ,
    L = 5.0,
    S = 1.0,
    venue = "Oakland Colosseum",
    state = "California",
    coordinates = Coordinates(
       lat = 37.7515541416315,
       lon = -122.20063165403968
)))

simulation.render_current_model(league_name)                            
```

### MLB moving Athletics to formally to Sacramento

```python
from sportwarz import Team, Coordinates

simulation.update_team(league_name, "Oakland A's", 
  Team(  
   name = "Sacramento A's" ,
   venue = "Sutter Health Park 2.0",
   state = "California",
   coordinates = Coordinates(
       lat =   38.58048,
       lon = -121.51352
)))

simulation.render_current_model(league_name)
```
