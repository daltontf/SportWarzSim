### MLS right now

```python
from sportwarz import Simulation

league_name = "MLS"

simulation  = Simulation()
simulation.add_league(league_name)
simulation.render_current_model(league_name)                            
```

### MLS moving Whitecaps to Las Vegas

```python
from sportwarz import Team, Coordinates

simulation.update_team(league_name, "Vancouver Whitecaps FC", 
  Team(  
    name = "Real Inter Las Vegas United FC" ,
    venue = "Some Field",
    state = "Nevada",
    color = "yellow",
    coordinates = Coordinates(
        lat = 36.102504,
        lon = -115.178729
    ))) 

simulation.render_current_model(league_name)
```
