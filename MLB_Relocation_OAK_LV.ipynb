```python
from sportwarz import LeaguesModel

leagues_model = LeaguesModel()

leagues_model.load_counties_geojson()
```

```python
league_name = "MLB"

leagues_model.load_leagues([league_name])
```

### MLB right now

```python
leagues_model.heatmap_counties(league_name)
leagues_model.render_map(league_name)
```

### Put Athletics back to Oakland

```python
# Complete move of A's to Vegas
from sportwarz import Team, Coordinates

team_OAK = Team(
    name = "Oakland A's" ,
    L = 5.0,
    S = 1.0,
    venue = "Oakland Col",
    state = "California",
    coordinates = Coordinates(
        lat = 37.7515541416315,
        lon = -122.20063165403968
))

leagues_oakland = leagues_model.copy_with_just_league(league_name)

leagues_oakland.update_team(league_name, "Athletics", team_OAK)

leagues_oakland.heatmap_counties(league_name) 

leagues_oakland.render_map(league_name)
```

### MLB after A's complete move to Vegas

```python
# Complete move of A's to Vegas
from sportwarz import Team, Coordinates

team_LVAs = Team(
    name = "Las Vegas A's" ,
    L = 5.0,
    S = 1.0,
    venue = "Fisher's Armadillo",
    state = "Nevada",
    coordinates = Coordinates(
        lat = 36.099444,
        lon = -115.170000
))

leagues_vegas = leagues_oakland.copy_with_just_league(league_name)

leagues_vegas.update_team(league_name, "Oakland A's", team_LVAs)

leagues_vegas.heatmap_counties(league_name) 

leagues_vegas.render_map(league_name)
```

```python
leagues_oakland.show_pre_post_merged_results(
    league_name,
    leagues_vegas
)
```
