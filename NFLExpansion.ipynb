```python
from sportwarz import Team, Coordinates, league_teams_sums, add_team, reset_county_styles, heatmap_counties

league_name = "NFL"

team_STL = Team(
    name = "St. Louis Stallions" ,
    L = 3.0,
    S = 3.0,
    venue = "National Car Rental Stadium",
    state = "Missouri",
    color = "plum",
    coordinates = Coordinates(
        lat =  38.63911966902621,
        lon = -90.182295076939940
))

leagues_NFL_wSTL = add_team(co_data_frame, league_name, [team_STL])

reset_county_styles(counties_geojson)
heatmap_counties(leagues_NFL_wSTL, counties_geojson, league_name) 
render_map(leagues_NFL_wSTL, league_name, counties_geojson, co_data_frame)
```

```python
import pandas as pd

pd.set_option('display.float_format', lambda x: '%.0f' % x)

def show_pre_post_merged_results(league_before, league_after):
    pre_sums = league_teams_sums(league_before)
    post_sums = league_teams_sums(league_after)
    merged = pd.merge(pre_sums, post_sums, on='team_name', how='right')
    merged["share_population_x"].fillna(0, inplace = True) 
    merged["share_population_delta"] = merged["share_population_y"] - merged["share_population_x"]
    print(merged)
    print(f"Sums\t: {merged["share_population_x"].sum()}\t{merged["share_population_y"].sum()}")
    print(f"Sum Delta = {merged["share_population_y"].sum() - merged["share_population_x"].sum()}") 

show_pre_post_merged_results(
    leagues[league_name],
    leagues_NFL_wSTL[league_name], 
)
```
