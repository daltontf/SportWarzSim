### MLB right now

```python
from sportwarz import Simulation

league_name = "MLB"

simulation  = Simulation()
simulation.add_league(league_name)
simulation.render_current_model(league_name)                            
```

### MLB moving White Sox to Nashville

```python
from sportwarz import Team, Coordinates

simulation.relocate_team(league_name, "Chicago White Sox", 
  Team(  
    name = "Nashville Stars" ,
    color = "black",
    venue = "Nashville Park",
    state = "Tennessee",
    coordinates = Coordinates(
        lat = 36.16222197198865,
        lon = -86.76564582516838
)))

simulation.render_current_model(league_name)
```
